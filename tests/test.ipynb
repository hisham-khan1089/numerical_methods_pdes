```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import time

# solving the nonhomogeneous case of the 2D heat equation, involving the addition
# of a function f(x, y, t)

start = time.perf_counter()

L = 50 # side length of square plate
iterations = 100

k = 1.0
delta_x = 1

delta_t = 0.25
s = (k * delta_t) / (delta_x ** 2)

# generate square plate for each iteration
u = np.empty((iterations, L, L))
u.fill(0)

def f(x, y, t):
    return 50*np.e**-((x-25)**2 + (y-25)**2)

for i in range(L): # note that range() should be taking L/delta_x, assuming that it returns an integer
    for j in range(L):
        u[0][j][i] = f(i*delta_x, j*delta_x, 0)
        

x = np.arange(0, L, step = delta_x)
y = np.arange(0, L, step = delta_x)
X, Y = np.meshgrid(x, y)

fig, ax = plt.subplots(subplot_kw=dict(projection='3d'))
ax.plot_surface(X, Y, u[0], cmap=plt.cm.viridis)
```
